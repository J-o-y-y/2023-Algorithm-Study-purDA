```python
# 학생 5명의 시험 점수를 입력받아 합계와 평균을 출력.

print('학생 그룹 점수의 합계와 평균을 구합니다.')

score1 = int(input('1번 학생의 점수를 입력하세요.: '))
score2 = int(input('2번 학생의 점수를 입력하세요.: '))
score3 = int(input('3번 학생의 점수를 입력하세요.: '))
score4 = int(input('4번 학생의 점수를 입력하세요.: '))
score5 = int(input('5번 학생의 점수를 입력하세요.: '))

total = score1 + score2 + score3 + score4 + score5
average = total / 5

print(f'합계는 {total}점입니다.')
print(f'평균은 {average}점입니다.')
```

```python
# 학생 5명의 시험 점수를 입력받아 합계와 평균을 출력하기

# 사용자에게 메시지를 출력합니다.
print('학생 그룹 점수의 합계와 평균을 구합니다.')

# 1번 학생의 점수를 입력받고, 정수형으로 변환하여 score1에 저장합니다.
score1 = int(input('1번 학생의 점수를 입력하세요.: '))

# 2번 학생의 점수를 입력받고, 정수형으로 변환하여 score2에 저장합니다.
score2 = int(input('2번 학생의 점수를 입력하세요.: '))

# 3번 학생의 점수를 입력받고, 정수형으로 변환하여 score3에 저장합니다.
score3 = int(input('3번 학생의 점수를 입력하세요.: '))

# 4번 학생의 점수를 입력받고, 정수형으로 변환하여 score4에 저장합니다.
score4 = int(input('4번 학생의 점수를 입력하세요.: '))

# 5번 학생의 점수를 입력받고, 정수형으로 변환하여 score5에 저장합니다.
score5 = int(input('5번 학생의 점수를 입력하세요.: '))

# 입력받은 5명 학생의 점수를 모두 더하여 total에 저장합니다.
total = score1 + score2 + score3 + score4 + score5

# 전체 점수의 합을 5로 나누어 평균을 계산하고, average에 저장합니다.
average = total / 5

# 합계와 평균을 화면에 출력합니다.
print(f'합계는 {total}점입니다.')
print(f'평균은 {average}점입니다.')
```
