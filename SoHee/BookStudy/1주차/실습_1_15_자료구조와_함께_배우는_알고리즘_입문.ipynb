# *를 n개 출력하되 W개마다 줄바꿈을 하는 프로그램

- 실습 1-14는 for 문을 반복할 때마다 if 문을 실행하므로 효율적X.
- 이를 실습 1-15와같이 개선.

```python
# 15

print('* 출력')
n = int(input('몇 개 출력?: '))
w = int(input('몇 개마다 줄바꿈?: '))

for _ in range(n // w):
    print('*' * w)

rest = n % w

if rest:
    print('*' * rest)
```

```python
# 별표 출력 프로그램입니다.

# 사용자로부터 출력할 별표 개수(n)를 입력받습니다.
n = int(input('몇 개 출력?: '))

# 사용자로부터 몇 개마다 줄바꾸기를 할지를 입력받습니다.
w = int(input('몇 개마다 줄바꿈?: '))

# n을 w로 나눈 몫만큼 별표를 출력하고 줄바꾸기를 수행합니다.
for _ in range(n // w):
    print('*' * w)

# n을 w로 나눈 나머지를 계산합니다.
rest = n % w

# 나머지가 0이 아닌 경우, 나머지 개수만큼 별표를 출력합니다.
if rest:
    print('*' * rest)
```

```python
# 14

print('*출력')
n = int(input('몇 개 출력?:'))
w= int(input('몇개마다 줄바꿈?:'))

for i in range(n):
  print('*',end='')
  if i % w == w-1:
    print()
if n % w:
  print()
```

```python
# '*' 출력 프로그램
print('*출력')

# 사용자로부터 몇 개의 '*'을 출력할지 입력을 받습니다.
n = int(input('몇 개 출력?: '))

# 몇 개마다 줄바꾸기를 할지 입력을 받습니다.
w = int(input('몇 개마다 줄바꿈?: '))

# 0부터 n-1까지 반복합니다.
for i in range(n):
    # '*' 문자를 출력하되, 줄바꿈 없이 출력하기 위해 'end' 매개변수를 사용합니다.
    print('*', end='')

    # 만약 i가 w-1로 나누어떨어지면, w개 마다 줄바꾸기를 수행합니다.
    if i % w == w - 1:
        print()

# 만약 n이 w로 나누어 떨어지지 않으면, 나머지 '*'을 출력하기 위해 줄바꾸기를 수행합니다.
if n % w:
    print()
```

# 14와 15의 차이점과 각각의 장단점.


| 특성                   | 코드 14                 | 코드 15                    |
|-----------------------|-----------------------|--------------------------|
| 출력 방식               | `*`를 반복 출력하고 `if`문을 사용하여 줄바꿈 처리 | `*`를 `n // w`번 출력하고 나머지를 처리하여 줄바꾸기 |
| 가독성                  | 상대적으로 낮음            | 상대적으로 높음            |
| 효율성                  | 상대적으로 낮음            | 상대적으로 높음            |
| 유지보수성              | 낮음                       | 높음                       |

**코드 14:**
- 출력 방식은 `*`를 반복 출력하고 `if`문을 사용하여 `w` 개마다 줄바꿈을 처리합니다. 이로 인해 코드가 상대적으로 복잡합니다.
- 가독성이 낮으며, 코드의 이해가 어려울 수 있습니다.
- 효율성 면에서도 낮습니다. 매 반복에서 `if`문을 실행하여 줄바꾸기를 확인하므로 비효율적입니다.
- 유지보수성이 낮아 코드 수정 및 확장이 어려울 수 있습니다.

**코드 15:**
- 출력 방식은 `*`를 `n // w`번 출력하고 나머지를 처리하여 줄바꾸기를 수행합니다. 이로 인해 코드가 간결하고 직관적입니다.
- 가독성이 높으며, 코드를 이해하기 쉽습니다.
- 효율성 면에서 우수합니다. 나머지 연산(`%`)을 사용하여 나머지 부분을 처리하므로 더 효율적입니다.
- 유지보수성이 높아 코드 수정이 쉽고, 필요한 경우 주석을 추가하기가 편리합니다.

요약하면, 코드 15는 코드 14에 비해 가독성이 높고 효율적으로 동작하는 간결한 방식으로 출력하고 줄바뾸기를 처리합니다. 이로 인해 코드를 이해하고 유지보수하는 데 더 용이하며, 일반적으로 더 좋은 선택입니다.
